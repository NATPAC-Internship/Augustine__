```python
import pandas as pd
import matplotlib.pyplot as plt



# Assuming the columns are named with years from 2015 to 2028
years = list(range(2015, 2029))
sums_with_offset = []

# Define the specific adjustments for each year
adjustments = {
    2025: -400000,
    2026: 40000000,
    2027: 20000000,
    2028: 50000000
}

# Calculate the sum of values for each year column and apply the adjustments
for year in years:
    sum_values = df[str(year)].sum()
    if year in adjustments:
        sum_values += adjustments[year]
    sums_with_offset.append(sum_values)

# Plotting the sum of values for each year with the specified adjustments
plt.figure(figsize=(10, 6))  # Optional: set the figure size

# Plot the calculated sums with the adjustments
plt.plot(years, sums_with_offset, marker='o', linestyle='-', label='sums of values')

# Adding titles and labels
plt.title('Sum of Values from 2015 to 2028 in Allapuzha')
plt.xlabel('Year')
plt.ylabel('Sum of Values')

# Adding a legend
plt.legend()

# Display the plot
plt.grid(True)  # Optional: adds a grid to the plot
plt.show()
```
